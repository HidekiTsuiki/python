## パズルの解を数える
<img src="s11.png" width="200">
手前左下から，箱に  0,1,2,3,4,5,6,7 の番号をつける。
また，H と T の向きを，
   T の向き 小さい三角がこの番号のどちらを頂点が向いているか 0,1,2,3,4,5,6,7
   H の向き それ + 4  (8,9,10,11)
で表す。


```python
# 手前左下から，箱に  0,1,2,3,4,5,6,7 の番号をつける。
# また，H と T の向きを，
#   T の向き： 小さい三角がこの番号のどちらを頂点が向いているか 0,1,2,3,4,5,6,7
#   H の向き：尖った下頂点 + 4  (8,9,10,11)
#で表す。
#possible:  それぞれの箱に，どのピースをどの向きに置けるか。

possibles = [[0], [1,7], [2,7], [3,5,6,8], 
                   [4,7],[3,5,6,8],[3,5,6,8], [0,1,2,4,7,9,10,11]]
assign = [0]*8
count = 0;
acount = 0;
solution = []
asolution = []

def find(n, countT, countH):
    global acount, count;
    global assign;
    if (n == 8):
        count= count+1;
        if(not (mirror(assign) in solution)):
            acount = acount + 1;
            asolution.append(assign.copy());
        solution.append(assign.copy());

        return
    for p in possibles[n]:
        if (p < 8 and countT < 5):
            assign[n]= p
            find(n+1, countT+1, countH)
        if (p >= 8 and countH < 3):
            assign[n]= p
            find(n+1, countT, countH+1)
find(0,0,0)

mirrors = [0,2,1,3,4,6,5,7,8,10,9,11]
def mirror(kai):
    out = [0]*8
    for i in range(8):
        out[i] = mirrors[kai[mirrors[i]]]
    return(out) 
print(count, acount)
```

```python
# 対称性のあるもの
u = [p for p in asolution if (mirror(p) == p)]
print(len(u))
u
```

```python
# 対称性のないもの
u = [p for p in asolution if (mirror(p) != p)]
print(len(u))
u
```

```python
# 1,2,4 は独立なので，７,7,7 に固定。しかも，対称性のないもの。
#これは，目で確認
u = [p for p in asolution if (mirror(p) != p) and p[1] == 7 and p[2] == 7 and p[4] == 7]
print(len(u))
u
```
