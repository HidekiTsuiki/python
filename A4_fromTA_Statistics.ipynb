# 第４回TAからの練習問題と解答例

## 問題1
20%の確率で当たるくじを10000回(1万回)引いた時の当たる回数について考える。  
以下の（ア）～（エ）に適切な数字を入れよ。  
有効数字に関しては、厳密に考えなくてもよい。
(もちろん、正規分布表をみてよい。)

・当たる回数の平均は（ア）回である。  
・1万回のうち、2040回以上当たる確率は約（イ）%である。    
・1万回のうち、（ウ）回以上当たる確率は約2.5%である。  
・1万回のうち、（エ）回以下しか当たらない確率は約2.5%である。  
したがって、当たる回数は95%の確率で（エ）回～（ウ）回の範囲に入っているとわかる。

## 問題1の解答例
くじを１回引いた時に当たる回数を表す確率変数を$X$とする。
$X$の確率分布は以下のようになる。

|  X |0|1|  計 |
| :----: |: ---- :|:----: |:----: |
|  p   | 0.8|0.2|1|

$X$と独立で$X$と同じ分布$X_{1},X_{2},X_{3},\dots,X_{10000}$をとってくる。  
$X_{1}+X_{2}+X_{3}+\dots+X_{10000}$が1万回引いた時の当たる回数を表している。  この分布が知りたい。  
中心極限定理より$X_{1}+X_{2}+X_{3}+\dots+X_{10000}$は正規分布で近似できる。  
以下、その正規分布の平均と分散を求める。
平均や分散の性質(問題掲載時の資料に記載)を繰り返し用いて、  
$E(X_{1}+X_{2}+X_{3}+\dots+X_{10000})=E(X_{1})+E(X_{2})+E(X_{3})\dots+E(X_{10000})=10000×E(X)=10000×0.2=2000$
$V(X_{1}+X_{2}+X_{3}+\dots+X_{10000})=V(X_{1})+V(X_{2})+V(X_{3})\dots+V(X_{10000})=10000×V(X)$  
ここで$V(X)=E(X^{2})-(E(X))^{2}$であったことを用いると、$V(X)=0.2-0.2×0.2=0.16$とわかり、
$V(X_{1}+X_{2}+X_{3}+\dots+X_{10000})=1600$とわかる。
（ア）$\cdots$2000

(参考)$X^{2}$の確率分布は

|  X^2 |0|1|  計 |
| :----: |: ---- :|:----: |:----: |
|  p   | 0.8|0.2|1|

であるから、$E(X^{2})=0.2$である。(参考終わり)

以下、$X_{1}+X_{2}+X_{3}+\dots+X_{10000}$を$Y$と置こう。
$Y$は平均2000、分散1600(標準偏差40)の正規分布であるから、正規化をすることで、
$$\frac{Y-2000}{40}$$は標準正規分布に従うことがわかる。  
$Y>2040$と$\frac{Y-2000}{40}>1$は同値である。  
標準正規分布において、1以上の値をとる確率は正規分布表を見ると、15.9%とわかる。  
よって、$Y>2040$となる確率も15.9%である。(イ)$\cdots$15.9%  
標準正規分布表を見ると、標準正規分布において1.96以上の値をとる確率は2.5%(2.5%の確率で1.96以上の値をとると同義)とわかる。  
$\frac{Y-2000}{40}>1.96$は、$Y>2078.4$と同値であるので、$Y$が2078以上の値をとる確率も約2.5%とわかる。(ウ)$\cdots$2078(2079)  
正規分布の対称性より、$Y$が1922以下の値をとる確率も約2.5%とわかる。（エ）$\cdots$1922(1921)

## 問題2
アームが２つの場合のバンディット問題を考える。  
今、一方のアームの当たる確率が40%で、もう一方のアームの当たる確率が60%であることが分かっている。  
ただし、どちらが60%で当たるアームなのかはわからない。  
これから１つのアームを引き続ける。(もう一方は一度も引かない。)  
引き続けているアームがどちらのアームなのかを99%以上の確率で当てるためには、
選んだアームから何回引けば十分か？

## 問題2の解答例
この問題には、統計学の重要な要素がたくさん詰まっているので、解けなかった方は必ず解答を読んでおきましょう。  
この問題をスラスラ解けて解答の数値も正しく導けた人は、統計学の基本がかなり身についていると思いますので自信を持ってください！！！

まず、考え方が少し独特で難しいかもしれません。  
例えば、100回引いて70回当たったらどうでしょう？60%のアームか40%のアームかと聞かれたら60%と答えますよね。  
では、100回引いて51回当たったらどうでしょう？  
60%か40%かどちらかから選べと言われたら、自信はないけど「どちらかと言われたら60%かな$\dots$」となりますよね。  
ここで「わかりません。」という答えはないわけです。  
視力検査の際、自信がない場合にすぐに「わかりません！」という方がいらっしゃいますが(ちなみに視力検査の際でも自信がなくても雰囲気でわかれば答えるというのが正しい測定方法です！)、統計的に予想する際は特に指定がない場合、**より可能性が高い方を選択します。**  (特に指定がない場合と書きましたが、例えば、90%以上の正解率が確保できない場合は回答しないものとするといった指定なども考えられはします。)  


つまり$n$回引いて$\frac{n}{2}$回以上当たれば60%のアームと予想し、$\frac{n}{2}$回以下しか当たらなければ40%のアームと予想することになります。
(60%と40%には対称性があるのでこのような議論が可能です。)
つまり、60%のアームを引き続けたとして、誤答するパターンは「$n$回引いた結果、当たった回数が$\frac{n}{2}$回未満だった」という場合です。その場合は「40%のアームだ」と回答し誤答となります。この確率を1%以下に抑えなければなりません。

60%で当たるアームを１回引いた時に当たる回数を表す確率変数を$X$とする。
$X$の確率分布は以下のようになる。

|  X |0|1|  計 |
| :----: |: ---- :|:----: |:----: |
|  p   | 0.4|0.6|1|

**引く回数を$n$**とする。
$X$と独立で$X$と同じ分布$X_{1},X_{2},X_{3},\dots,X_{n}$をとってくる。  
$X_{1}+X_{2}+X_{3}+\dots+X_{n}$が$n$回引いた時の当たる回数を表している。  
($n$が十分大きいと仮定して、)中心極限定理より$X_{1}+X_{2}+X_{3}+\dots+X_{n}$は正規分布で近似できる。…(★)  
以下、その正規分布の平均と分散を求める。
上に書いた性質を繰り返し用いて、  
$E(X_{1}+X_{2}+X_{3}+\dots+X_{n})=E(X_{1})+E(X_{2})+E(X_{3})\dots+E(X_{n})=n×E(X)=n×0.6=0.6n$
$V(X_{1}+X_{2}+X_{3}+\dots+X_{n})=V(X_{1})+V(X_{2})+V(X_{3})\dots+V(X_{n})=n×V(X)$  
ここで$V(X)=E(X^{2})-(E(X))^{2}$であったことを用いると、$V(X)=0.6-0.6×0.6=0.24$とわかり、
$V(X_{1}+X_{2}+X_{3}+\dots+X_{n})=0.24n$とわかる。

以下、$X_{1}+X_{2}+X_{3}+\dots+X_{n}$を$Y$と置こう。
$Y$は平均$0.6n$、分散$0.24n$(標準偏差$\sqrt{0.24n}$)の正規分布であるから、正規化をすることで、
$$\frac{Y-0.6n}{\sqrt{0.24n}}$$は標準正規分布に従うことがわかる。  
$Y<\frac{n}{2}$と$\frac{Y-0.6n}{\sqrt{0.24n}}<-\frac{0.1n}{\sqrt{0.24n}}$は同値である。

$Y<\frac{n}{2}$となる確率、つまり標準正規分布において$-\frac{0.1n}{\sqrt{0.24n}}$以下の値をとる確率が誤答してしまう確率なので、これが1%以下になるようにする。
標準正規分布表より、標準正規分布において、$-2.32$以下の値をとる確率が1%とわかるので、
$-\frac{0.1n}{\sqrt{0.24n}}=-2.32$の時に、誤答してしまう確率が1%となる。  
あとはこの方程式を解けばよい。(ここでの計算間違いが意外と起こりやすいので注意！！！)  
両辺を2乗して、$\frac{0.1×0.1×n}{0.24}=2.32×2.32$,  
$n=129.1776$  
この$n$は十分に大きいので(★)が正当化できる。
よって、129(130)回程度、同じアームを引き続けることで、引いているアームが60%のアームの場合、99%の確率で60%のアームであることを当てられる。対称性より、引き続けているアームが40%のアームの場合、99%の確率で40%のアームであることを当てられる。  
つまり、129(130)回程度、同じアームを引き続けることで、引き続けているアームがどちらのアームなのかを99%で当てることができる。

**この問題で気を付けてほしいのが99%の扱いです。**
「99%で○○以上○○以下の値をとる。」という場合、○○以上の確率が0.5%,○○以下の確率が0.5%になるようにしなければなりません。
「99%で○○以上の値をとる。」という場合、○○以下の確率が1%になるようにしなければなりません。
片側なのか両側なのか確認をし、使う数値を間違えないようにしましょう！！！

$$
である。計算は大変であるが、高校の数Ｂで習った数列の和の公式を使うなどすると値を求めることができ、

## 問題3の解答例　離散分布の場合
$Y$が$0.01n$以下であることと、$X_{1},X_{2},X_{3}$が全て$0.01n$以下であることは同値である。  
$X$が$0.01n$以下である確率は$0.01n$なので、$X_{1},X_{2},X_{3}$の全てが$0.01n$以下である確率は$0.01n$の３乗である。  
したがって、$Y$が$0.01n$以下である確率は$0.01n$の３乗である。  
$Y$がちょうど$0.01n$である確率は、「$Y$が$0.01n$以下である確率」から「$Y$が$0.01(n-1)$以下である確率」を引いた値なので、${0.01n}^3-{0.01(n-1)}^3=0.01^{3}(n^3-(n-1)^3)=0.01^{3}(n^3-n^3+3n^2-3n+1)=0.01^{3}(3n^2-3n+1)$である。  

これで$Y$の確率分布が求まった。
平均の定義より、
$$
E(Y)=\sum_{n=1}^{100}(0.01n×0.01^{3}(3n^{2}-3n+1))
$$
である。計算は大変であるが、高校の数Ｂで習った数列の和の公式を使うなどすると値を求めることができ、
$$
E(Y)=\sum_{n=1}^{100}(0.01n×0.01^{3}(3n^{2}-3n+1))=0.01^{4}×\sum_{n=1}^{100}(3n^{3}-3n^{2}+n)\\
=0.01^{4}×\left(3×\left(\frac{100×101}{2}\right)^{2}-3×\left(\frac{100×101×201}{6}\right)+\frac{100×101}{2}\right)=0.754975
$$とわかる。

## 問題3の解答例　連続分布の場合

$Y≦k$の確率は$k^{3}$である。($0≦k≦1$)  
よって、$Y$の累積分布関数は$F(x)=x^{3}$である。  
$Y$の確率密度関数は$f(x)=F'(x)=3x^{2}$となる。  
よって、$Y$の期待値は、
$$\int_{0}^{1}xf(x)dx=\int_{0}^{1}x×3x^{2}dx=\int_{0}^{1}3x^{3}dx=\left[\frac{3}{4}x^{4}\right]_{0}^{1}=\frac{3}{4}=0.75$$
とわかる。  
連続分布で考えた方が計算自体は楽ですね。
