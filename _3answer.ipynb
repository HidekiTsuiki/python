```python
# 練習問題20
def prime (x):
    for i in range(2,x):
        if(x % i == 0):
            return False
    return True

prime(13)
```

```python
# 練習問題30
def numprime(x):
    c = 0
    for i in range(2,x+1):
        if(prime(i)):
            c += 1
    return c

numprime(10)
```

```python
# 練習問題40
import random
def pi():
    repeatnum = 1000000
    c = 0
    for _ in range(repeatnum):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        if(x**2 + y**2 < 1):
            c = c + 1
    return(4 * c / repeatnum)

pi()
    
```

```python
#練習問題50
def nextprime(n):
    while(True):
        n += 1
        if(prime(n)):
            return n
        
nextprime(101)        
```

```python
# 練習問題60
def repeat(f, n=2, x=0.5):
    for _ in range(n):
        x = f(x)
    return x

def square (x):
    return x * x
print(repeat(square))
print(repeat(square, x=1.2))
print(f"{repeat(square, 7, 1.2):f}")
```

```python
# 練習問題 70
def hoteishiki2(f, target=10000, n=1000):
    for i in range(n):
        if(f(i)==target):
            return i
    return -1
           
```

```python


# 練習問題112
def newton1(f, df, u=2,epsilon=0.00001):
    for i in range(1000):
        u1 = u - (f(u) / df(u)) ;
        if(abs(u1 - u) < epsilon):
#            print(i)
            break
        u = u1
    return u

y = newton1(lambda x: x**10 - 2, lambda x : 10 * x**9)
print(y, y**10, 2**0.1)
```

```python
# 練習問題115
def newton(f, df=0, u=2,epsilon=0.00001):
    if(df == 0): #関数 df が与えられなかった
        df = lambda x: (f(x+ epsilon) - f(x)) / epsilon
    return newton1(f, df, u, epsilon)

y = newton(lambda x: x**10 - 2)
print(y, y**10, 2**0.1)
```

```python
# 練習問題100-2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def newtonarea(f, fro = -2, to = 2, num = 200, epsilon=0.0001 ):
    x = np.linspace(fro, to, num);
    df = lambda x: (f(x+ epsilon) - f(x)) / epsilon
    plt.plot(x, f(x))
    plt.plot(x, df(x))
#    for u in x:  # 実は，x は，from から to までを num で区切った数のリスト。次回習う
                 # それまでは，for i in range(num) で，u = fro + i * (to - fro)/num とすべき
    for i in range(num):
        u = fro + i * (to - fro)/num
        y = newton(f, df, u)
        if(y < -1.5): 
            plt.scatter(u, 0, s = 0.1, c = "yellow")  # 収束しなかった？
        if(y < -0.5):
            plt.scatter(u, 0, s = 0.1, c = "red")
        elif(y < 0.5):
            plt.scatter(u, 0, s = 0.1, c = "green")
        elif(y < 1.5):
            plt.scatter(u, 0, s = 0.1, c = "blue")
        else:
            plt.scatter(u, 0, s = 0.1, c = "yellow")　 # 収束しなかった？

    plt.show()

newtonarea(lambda x: x * (x-1) * (x + 1))
```
