```python
# 練習問題60
def repeat(f, n=2, x=0.5):
    for _ in range(n):
        x = f(x)
    return x

def square (x):
    return x * x
print(repeat(square))
print(repeat(square, x=1.2))
print(f"{repeat(square, 7, 1.2):f}")
```

```python
def at_10(f):
    return f(10)

def square(x):
    return x*x

def cube(x):
    return x*x*x

print(at_10(square), at_10(cube))
print(at_10(float))
```

```python
# 練習問題 70
def hoteishiki2(f, target=10000, n=1000):
    for i in range(n):
        if(f(i)==target):
            return i
    return -1
           
```

```python
print(hoteishiki2(square))
print(hoteishiki2(cube, 1000000))
```

```python
#練習問題 80
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def showfun(f, s=0, e=1, d=1000):
    x = np.linspace(s, e, d)
    plt.plot(x, f(x))
    plt.show()
#    plt.savefig(str(f)+"fig.jpg")
```

```python
showfun(square,-1)
```

```python
showfun(np.sin, 0, 10*np.pi, 1000)
```

```python
#練習問題90
def repeatfun(f, n):
    return(lambda x: repeat(f, n, x))
```

```python
# 練習問題100-1
f = lambda x: 2.9*x*(1-x)
g = repeatfun(f, 2)
h = repeatfun(f, 3)

showfun(f)
showfun(g)
showfun(h)
```

```python
# 練習問題31
f = lambda x: 2.9*x*(1-x)
x = np.linspace(0, 1, 100);
for i in range(10):
    showfun(repeatfun(f, i))
```

```python
# 練習問題100-2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: 2.9*x*(1-x)
x = np.linspace(0, 1, 100);
for i in range(100):
    plt.plot(x, repeatfun(f, i)(x))
plt.show()
```
