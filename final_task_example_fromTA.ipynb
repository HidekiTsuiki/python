# 5．Alphazeroなどの機械学習

Alphazeroをはじめとする人工知能を用いて、ゲームAIを作ってみよう。
Alphazeroであれば、書籍『AlphaZero深層学習・強化学習・探索 : 人工知能プログラミング実践入門』にソースコードも載っており参考になる。
Alphazeroに限らず他のアルゴリズムや自分で考えたアルゴリズムで学習してみるのも面白い。

Alphazeroの場合、正しく動作していたとしても我々のＰＣでは9×9の囲碁や8×8オセロは学習ができない。(重要！！)  
tic-tac-toe(三並べ)や4×4オセロで正しく学習できるか試してみよう。
tic-tac-toeであれば、多少悪いパラメータ設定にしていても学習できるはずである。
もう少し困難なゲーム（6×6オセロなど）の場合は、ハイパーパラメータを上手く定めることができれば学習が進むはずである。
上手くいかなかった場合でもその過程や考察を発表してもらえれば十分である。
